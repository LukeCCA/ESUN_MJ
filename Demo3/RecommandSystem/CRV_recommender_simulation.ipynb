```python
import random as rd
import numpy as np
import pandas as pd
```

```python
Offer_list=[0,1]
Pref=[0.75,0.25] #offer給0的話有75%機率回應,給1的話有25%機率回應
response=np.zeros([1000,100,2])#1000人,每人來訪100次,第三個維度0為是否有回應,1為給的offer
```

```python
#第一次offer一律給1
#正確offer應為0,可觀察過程如何調整到正確offer
#第二次起根據第一次到最近一次每次的回應狀況,計算哪一個offer的回應比率較高則給該offer,

def offer_rule(i,j,response):
    if j==0:
        response[i,j,1]=1 #cold start,預設給1
    if j!=0:    
        if response[i,j-1,1]==0:
            response_A[i,j-1,0]=response[i,j-1,0]
            response_A[i,j-1,1]=1          
        elif response[i,j-1,1]==1:
            response_B[i,j-1]=response[i,j-1,0]
            response_B[i,j-1,1]=1
            
        #若沒有給過該offer,回應比率當作0
        if sum(response_A[i,:j,1])!=0:
            ratio_A=sum(response_A[i,:j,0])/sum(response_A[i,:j,1])
        else:
            ratio_A=0
        if sum(response_B[i,:j,1])!=0:
            ratio_B=sum(response_B[i,:j,0])/sum(response_B[i,:j,1])
        else:
            ratio_B=0
            
        #若兩offer回應比率相同,給跟上次不同的offer
        if ratio_A > ratio_B:
            response[i,j,1]=0
        elif ratio_A < ratio_B:
            response[i,j,1]=1
        elif ratio_A == ratio_B and response[i,j-1,1]==1:
            response[i,j,1]=0
        elif ratio_A == ratio_B and response[i,j-1,1]==0:
            response[i,j,1]=1
```

```python
response_A=np.zeros([1000,100,2])
response_B=np.zeros([1000,100,2])
for i in range(1000):
    for j in range(100):
        offer_rule(i,j,response)
        if response[i,j,1]==0:
            threshold=Pref[0]
        elif response[i,j,1]==1:
            threshold=Pref[1]
        if np.random.rand()<threshold:
            response[i,j,0]=1
        else:
            response[i,j,0]=0
```
